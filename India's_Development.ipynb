````
# 4. Extract India-Specific Data and relevant year columns (2000-2024)

def get_year_columns(df):
    year_cols = []
    for col in df.columns:
        if isinstance(col, str) and ('[YR' in col or col.isdigit()):
            try:
                year = int(col.split(' ')[0]) if '[YR' in col else int(col)
                if 2000 <= year <= 2024:
                    year_cols.append(col)
            except ValueError:
                pass
    return year_cols

hdi_year_cols = get_year_columns(hdi_df)
public_sector_year_cols = get_year_columns(public_sector_df)
wdi_year_cols = get_year_columns(wdi_df)
wgi_year_cols = get_year_columns(wgi_df)

# Define the columns to keep for filtering
hdi_cols_to_keep = ['Country Name'] + hdi_year_cols + ['Value'] # Keep HDI value (named 'Value' in hdi_df)
public_sector_cols_to_keep = ['Country Name', 'Series Name', 'Series Code'] + public_sector_year_cols
wdi_cols_to_keep = ['Country Name', 'Indicator Name', 'Indicator Code'] + wdi_year_cols
wgi_cols_to_keep = ['Country Name', 'Series Name', 'Series Code'] + wgi_year_cols


# Filter for India and select relevant columns
# Assuming the header and column names are now correctly set

# Strip whitespace from 'Country Name' column and the filter string 'India'
hdi_df['Country Name'] = hdi_df['Country Name'].astype(str).str.strip()
public_sector_df['Country Name'] = public_sector_df['Country Name'].astype(str).str.strip()
wdi_df['Country Name'] = wdi_df['Country Name'].astype(str).str.strip()
wgi_df['Country Name'] = wgi_df['Country Name'].astype(str).str.strip()

country_filter = 'India'.strip()

print("\n--- Columns in hdi_df before filtering ---")
print(hdi_df.columns.tolist())
india_hdi_df = hdi_df[hdi_df['Country Name'] == country_filter][hdi_cols_to_keep]


print("\n--- Columns in public_sector_df before filtering ---")
print(public_sector_df.columns.tolist())
try:
    india_public_sector_df = public_sector_df[public_sector_df['Country Name'] == country_filter][public_sector_cols_to_keep]
except KeyError as e:
    print(f"Error filtering public_sector_df: {e}")
    print("Available columns in public_sector_df:", public_sector_df.columns.tolist())

print("\n--- Columns in wdi_df before filtering ---")
print(wdi_df.columns.tolist())
try:
    india_wdi_df = wdi_df[wdi_df['Country Name'] == country_filter][wdi_cols_to_keep]
except KeyError as e:
    print(f"Error filtering wdi_df: {e}")
    print("Available columns in wdi_df:", wdi_df.columns.tolist())

print("\n--- Columns in wgi_df before filtering ---")
print(wgi_df.columns.tolist())
try:
    india_wgi_df = wgi_df[wgi_df['Country Name'] == country_filter][wgi_cols_to_keep]
except KeyError as e:
    print(f"Error filtering wgi_df: {e}")
    print("Available columns in wgi_df:", wgi_df.columns.tolist())


print("\n--- India HDI DataFrame Head ---")
display(india_hdi_df.head())
print("\n--- India Public Sector Indicators DataFrame Head ---")
display(india_public_sector_df.head())
print("\n--- India World Development Indicators (WDI) DataFrame Head ---")
display(india_wdi_df.head())
print("\n--- India World Bank Governance Indicators (WGI) DataFrame Head ---")
display(india_wgi_df.head())
````

### Step 3: Merging and Structuring the Data

**Merge DataFrames**

````
# Function to clean year columns and unpivot the dataframe
def clean_and_melt_df(df, id_vars, year_cols, value_name='Value'):
    # Rename year columns to just the year (e.g., '2000 [YR2000]' to '2000')
    rename_dict = {col: col.split(' ')[0] for col in year_cols}
    df = df.rename(columns=rename_dict)

    # Check if the value_name conflicts with an existing column and rename if necessary
    original_value_col = None
    if value_name in df.columns and value_name not in id_vars:
        original_value_col = value_name
        temp_value_name = f"{value_name}_temp"
        df = df.rename(columns={original_value_col: temp_value_name})
        # Update value_vars if the original value column was in it
        if original_value_col in [col.split(' ')[0] for col in year_cols]:
             year_cols = [temp_value_name if x == original_value_col else x for x in year_cols]


    # Melt the dataframe to long format
    # Exclude the original value column from melting if it was renamed
    cols_to_melt = [col.split(' ')[0] for col in year_cols]
    if original_value_col:
        cols_to_melt = [col for col in cols_to_melt if col != original_value_col]

    melted_df = df.melt(id_vars=id_vars,
                        value_vars=cols_to_melt,
                        var_name='Year',
                        value_name=value_name)

    # If the original value column was renamed, add it back to the melted dataframe
    if original_value_col and temp_value_name in df.columns:
         # Merge the original value column based on id_vars
         value_col_df = df[id_vars + [temp_value_name]].rename(columns={temp_value_name: original_value_col})
         melted_df = pd.merge(melted_df, value_col_df, on=id_vars, how='left')


    # Convert 'Year' column to integer
    melted_df['Year'] = pd.to_numeric(melted_df['Year'], errors='coerce')
    return melted_df

# Clean and melt each dataframe
id_vars_hdi = ['Country Name']
india_hdi_melted = clean_and_melt_df(india_hdi_df.copy(), id_vars_hdi, hdi_year_cols, value_name='Value')
india_hdi_melted = india_hdi_melted.rename(columns={'Value': 'Human Development Index (HDI)'}) # Rename Value column


id_vars_public_sector = ['Country Name', 'Series Name', 'Series Code']
india_public_sector_melted = clean_and_melt_df(india_public_sector_df.copy(), id_vars_public_sector, public_sector_year_cols)
india_public_sector_melted = india_public_sector_melted.rename(columns={'Value': 'Indicator Value'}) # Rename Value column


id_vars_wdi = ['Country Name', 'Indicator Name', 'Indicator Code']
india_wdi_melted = clean_and_melt_df(india_wdi_df.copy(), id_vars_wdi, wdi_year_cols)
india_wdi_melted = india_wdi_melted.rename(columns={'Value': 'Indicator Value'}) # Rename Value column


id_vars_wgi = ['Country Name', 'Series Name', 'Series Code']
india_wgi_melted = clean_and_melt_df(india_wgi_df.copy(), id_vars_wgi, wgi_year_cols)
india_wgi_melted = india_wgi_melted.rename(columns={'Value': 'Indicator Value'}) # Rename Value column


# Merge the dataframes
merged_df = pd.merge(india_wdi_melted, india_hdi_melted, on=['Country Name', 'Year'], how='left')

# Combine Public Sector and WGI melted dataframes as they have similar structure (Series Name/Code)
combined_public_governance = pd.concat([india_public_sector_melted, india_wgi_melted]).drop_duplicates()

# combined dataframe from all melted data, adding a source column to distinguish.
india_hdi_melted['Source'] = 'HDI'
india_wdi_melted['Source'] = 'WDI'
india_public_sector_melted['Source'] = 'Public Sector'
india_wgi_melted['Source'] = 'WGI'

# Standardize column names before concatenation.
india_wdi_melted = india_wdi_melted.rename(columns={'Indicator Name': 'Series Name', 'Indicator Code': 'Series Code'})

india_hdi_melted['Series Name'] = 'Human Development Index (HDI)'
india_hdi_melted['Series Code'] = 'HDI.Value'
india_hdi_melted = india_hdi_melted.rename(columns={'Human Development Index (HDI)': 'Indicator Value'}) # Rename the HDI value column

# concatenate all melted dataframes
all_india_melted = pd.concat([india_hdi_melted, india_wdi_melted, india_public_sector_melted, india_wgi_melted], ignore_index=True)

# Drop duplicate rows from concatenating WGI and Public Sector if they overlap
all_india_melted = all_india_melted.drop_duplicates(subset=['Country Name', 'Year', 'Series Name', 'Series Code', 'Indicator Value', 'Source'])

# Display the head of the combined dataframe
print("\n--- Combined India Data DataFrame Head ---")
display(all_india_melted.head())

# Display info of the combined dataframe
print("\n--- Combined India Data DataFrame Info ---")
all_india_melted.info()
````
